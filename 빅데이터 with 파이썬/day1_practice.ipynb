```python
import numpy as np
import pandas as pd
```

```python
titanic = pd.read_csv('/content/titanic.csv')

titanic
```

```python
fe = titanic.loc[titanic['Sex'] == 'female']
m = titanic.loc[titanic['Sex'] == 'male']
```

```python
fe
```

```python
m
```

```python
fe_n = fe.loc[fe['Pclass'] == 1].count()
m_n = m.loc[m['Pclass'] == 1].count()

print(fe_n, m_n)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(1)
plt.ylim([0, 500])
plt.hist(fe['Pclass'])
plt.figure(2)
plt.ylim([0, 500])
plt.hist(m['Pclass'])

plt.show()
```

```python
s_n = titanic.loc[titanic['Survived'] == 1]
d_n = titanic.loc[titanic['Survived'] == 0]
```

```python
print(s_n['PassengerId'].sum())
print(d_n['PassengerId'].sum())
```

```python
plt.figure(1)
plt.ylim([0, 100])
plt.hist(s_n['PassengerId'])
plt.figure(2)
plt.ylim([0, 100])
plt.hist(d_n['PassengerId'])

plt.show()
```
