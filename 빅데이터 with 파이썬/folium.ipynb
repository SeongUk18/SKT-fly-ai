```python
import pandas as pd

# 대학교 리스트를 데이터프레임 변환
df = pd.read_excel('/content/seoulUniv.xlsx')
df.head(5)
```

```python
seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=11, width = 750, height= 500)
# 대학교 위치정보를 Marker로 표시
for name, lat, lng in zip(df['Name'], df['위도'], df['경도']):
 folium.Marker([lat, lng], popup=name).add_to(seoul_map)
seoul_map
```

```python
for name, lat, lng in zip(df['Name'], df['위도'], df['경도']):
 folium.CircleMarker([lat, lng], popup=name,
                     radius=10, # 원의 반지름
                     color='brown', # 원의 둘레 색상
                     fill=True, # 기본값 False
                     fill_color='coral', # 원을 채우는 색
                     fill_opacity=0.7 # 투명도
                     ).add_to(seoul_map)
seoul_map
```
