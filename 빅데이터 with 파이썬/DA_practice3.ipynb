```python
# 한글 폰트 가져오기
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings(action='ignore')
path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf' # 나눔 고딕
font_name = fm.FontProperties(fname=path, size=10).get_name() # 기본 폰트 사이즈 : 10
plt.rc('font', family=font_name)
```

```python
import pandas as pd

data = pd.read_csv('/content/한국노인인력개발원_노인일자리 통합정보_20240528.csv', encoding='cp949')

data.head(15)
```

```python
data.info()
```

```python
grouped_data = data.groupby('관활시도명').agg({
    '사업유형': lambda x: list(x),
    '목표일자리수': 'sum',
}).reset_index()


grouped_data.info()
```

```python
grouped_data['사업유형_갯수'] = grouped_data['사업유형'].apply(len)
grouped_data
```

```python
sorted_grouped_data = grouped_data.sort_values(by='목표일자리수', ascending=False)
```

```python
sorted_grouped_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Create a bar plot
sns.barplot(x='목표일자리수', y='관활시도명', data=sorted_grouped_data, palette='viridis')

# Add titles and labels
plt.title('city per jobs')
plt.xlabel('jobs')
plt.ylabel('city')

# Show the plot
plt.show()
```

```python
# 한글 폰트 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font_prop = fm.FontProperties(fname=font_path)
plt.rc('font', family=font_prop.get_name())

# 시각화
sns.set(style="whitegrid")

# Figure 생성
fig, ax1 = plt.subplots(figsize=(14, 7))

# 첫 번째 그래프: '사업유형_갯수'
sns.barplot(x='관활시도명', y='사업유형_갯수', data=grouped_data, ax=ax1, palette='Blues', ci=None)
ax1.set_ylabel('사업유형 갯수')
ax1.set_xlabel('관활시도명')
ax1.set_title('지역별 사업유형 갯수 및 목표일자리수')

# 두 번째 그래프: '목표일자리수' (y축 맞추기)
ax2 = ax1.twinx()
sns.lineplot(x='관활시도명', y='목표일자리수', data=grouped_data, ax=ax2, color='red', marker='o')
ax2.set_ylabel('목표일자리수')

# 그래프 표시
plt.show()
```

```python
# 데이터 준비
melted_data = grouped_data.melt(id_vars='관활시도명', value_vars=['사업유형_갯수', '목표일자리수'],
                                var_name='지표', value_name='값')

# 시각화
sns.set(style="whitegrid")
plt.figure(figsize=(14, 7))

# 막대 그래프
sns.barplot(x='관활시도명', y='값', hue='지표', data=melted_data, palette='pastel')

# 그래프 설정
plt.title('지역별 사업유형 갯수 및 목표일자리수')
plt.xlabel('관활시도명')
plt.ylabel('값')
plt.legend(title='지표')

# 그래프 표시
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np

# 한글 폰트 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font_prop = fm.FontProperties(fname=font_path)
plt.rc('font', family=font_prop.get_name())

# 바의 위치 조정
n = len(grouped_data)
ind = np.arange(n)
width = 0.35

# Figure와 첫 번째 축 생성
fig, ax1 = plt.subplots(figsize=(14, 7))

# 첫 번째 축: '목표일자리수'
bar1 = ax1.bar(ind - width/2, sorted_grouped_data['목표일자리수'], width, label='목표일자리수', color='r', alpha=0.6)
ax1.set_xlabel('관활시도명', fontproperties=font_prop)
ax1.set_ylabel('목표일자리수', color='r', fontproperties=font_prop)
ax1.tick_params(axis='y', labelcolor='r')
ax1.set_xticks(ind)
ax1.set_xticklabels(sorted_grouped_data['관활시도명'], rotation=45, fontproperties=font_prop)

# 두 번째 축 생성 및 두 번째 막대: '사업유형 갯수'
ax2 = ax1.twinx()
bar2 = ax2.bar(ind + width/2, sorted_grouped_data['사업유형_갯수'], width, label='사업유형 갯수', color='b', alpha=0.6)
ax2.set_ylabel('사업유형 갯수', color='b', fontproperties=font_prop)
ax2.tick_params(axis='y', labelcolor='b')

# 범례 설정
fig.legend(loc='upper right', bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes, prop=font_prop)

# 그래프 제목 설정
plt.title('지역별 사업유형 갯수 및 목표일자리수', fontproperties=font_prop)

# 그래프 표시
plt.show()
```

```python
# 한글 폰트 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font_prop = fm.FontProperties(fname=font_path)
plt.rc('font', family=font_prop.get_name())


# 목표일자리수의 총합 계산
total_jobs = sorted_grouped_data['목표일자리수'].sum()

# 각 지역의 비율 계산
sorted_grouped_data['비율'] = sorted_grouped_data['목표일자리수'] / total_jobs
# sorted_grouped_data
# # 4% 이하인 지역을 "기타"로 합치기
others = sorted_grouped_data[sorted_grouped_data['비율'] <= 0.04].sum(numeric_only=True)
others['관활시도명'] = '기타'
sorted_grouped_data = sorted_grouped_data[sorted_grouped_data['비율'] > 0.04]

# "기타" 데이터 추가
sorted_grouped_data = pd.concat([sorted_grouped_data, pd.DataFrame([others])], ignore_index=True)

# 파이차트 생성
fig, ax = plt.subplots(figsize=(10, 7))
ax.pie(sorted_grouped_data['목표일자리수'], labels=sorted_grouped_data['관활시도명'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))

# 제목 설정
ax.set_title('지역별 목표일자리수 비율', fontproperties=font_prop)

# 그래프 표시
plt.show()
```

```python
# 목표일자리수와 사업유형 갯수의 총합 계산
total_jobs = sorted_grouped_data['목표일자리수'].sum()
total_business_types = sorted_grouped_data['사업유형_갯수'].sum()

# 각 지역의 비율 계산
sorted_grouped_data['목표일자리수_비율'] = sorted_grouped_data['목표일자리수'] / total_jobs
sorted_grouped_data['사업유형_갯수_비율'] = sorted_grouped_data['사업유형_갯수'] / total_business_types

# 4% 이하인 지역을 "기타"로 합치기 (목표일자리수)
others_jobs = sorted_grouped_data[sorted_grouped_data['목표일자리수_비율'] <= 0.04].sum(numeric_only=True)
others_jobs['관활시도명'] = '기타'
sorted_grouped_data = sorted_grouped_data[sorted_grouped_data['목표일자리수_비율'] > 0.04]
sorted_grouped_data = pd.concat([sorted_grouped_data, pd.DataFrame([others_jobs])], ignore_index=True)

# 4% 이하인 지역을 "기타"로 합치기 (사업유형 갯수)
others_business_types = sorted_grouped_data[sorted_grouped_data['사업유형_갯수_비율'] <= 0.04].sum(numeric_only=True)
others_business_types['관활시도명'] = '기타'
grouped_data_business_types = sorted_grouped_data[sorted_grouped_data['사업유형_갯수_비율'] > 0.04]
sorted_grouped_data = pd.concat([grouped_data_business_types, pd.DataFrame([others_business_types])], ignore_index=True)

# 파이차트 생성 (목표일자리수)
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

ax[0].pie(sorted_grouped_data['목표일자리수'], labels=sorted_grouped_data['관활시도명'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
ax[0].set_title('지역별 목표일자리수 비율', fontproperties=font_prop)

# 파이차트 생성 (사업유형 갯수)
ax[1].pie(sorted_grouped_data['사업유형_갯수'], labels=sorted_grouped_data['관활시도명'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
ax[1].set_title('지역별 사업유형 갯수 비율', fontproperties=font_prop)

# 그래프 표시
plt.show()
```
