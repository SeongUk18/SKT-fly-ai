```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import matplotlib.pyplot as plt


res = requests.get('https://search.naver.com/search.naver?sm=mtp_hty.top&where=m&query=%EC%98%81%ED%99%94')
soup = bs(res.content, 'html.parser')
result = soup.find_all(class_='this_text _text')
# print(result)
result2 = soup.find_all(class_='num')
data = []
for name, num in zip(result, result2):
  data.append((name.text, float(num.text)))
# data

df = pd.DataFrame(data, columns=['name', 'rating'])
# df

plt.rc('font', family='NanumBarunGothic')
plt.bar(df['name'], df['rating'])
fig = plt.gcf()
fig.set_size_inches(20,3)
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup

def main():
    url = 'http://cine21.com/movie/point/view/?movie_id=40434'
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        comments = soup.select('.comment')  # CSS   'comment'    

        for comment in comments:
            print(comment.get_text(strip=True))
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")

if __name__ == "__main__":
    main()
```

```python
from collections import Counter
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def get_comments():
    driver = webdriver.Chrome()

    try:
        driver.get('http://cine21.com/movie/point/view/?movie_id=40434')

        #    
        wait = WebDriverWait(driver, 10)
        all_data = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "comment")))

        comments_text = [data.text for data in all_data]  # WebElement  

    finally:
        driver.quit()

    return comments_text

def generate_wordcloud(words):
    wordcloud = WordCloud(font_path='malgun.ttf', background_color='white').generate_from_frequencies(words)

    plt.figure(figsize=(10, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

def main():
    comments = get_comments()

    words = []

    for comment in comments:
        words.extend(comment.split())

    #   
    counter = Counter(words)

    #   
    generate_wordcloud(counter)

if __name__ == "__main__":
    main()
```
