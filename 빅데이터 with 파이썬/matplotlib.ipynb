```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y)
plt.plot(x, z)
plt.title('플롯 그래프')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x + 0.5)
y3 = np.sin(x + 1)
plt.plot(x, y1, 'r--')
plt.plot(x, y2, 'b.')
plt.plot(x, y3, 'g-.')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x + 0.5)
y3 = np.sin(x + 1)
plt.plot(x, y1, 'r--', label='Sin A')
plt.plot(x, y2, 'b.', label='Sin B')
plt.plot(x, y3, 'g-.', label='Sin C')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
plt.style.use('ggplot')
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r-')
plt.subplot(2, 2, 2)
plt.plot(x, z, 'g--')
plt.subplot(2, 2, 3)
plt.plot(x, z, 'b-.')
fig = plt.gcf()
fig.set_size_inches(3,3)
plt.show()
```

```python
name = ['A','B','C','D']
y = [5. , 25. , 50. , 20.]
z = [8. , 20. , 25. , 6.]
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(12, 5)
ax[0].bar(name, y, color='#33aaff')
ax[1].bar(name, z, color=['r', 'g', 'b', 'k'])
plt.show()
```

```python
import pandas as pd

titanic = pd.read_csv('/content/titanic.csv')
```

```python
titanic.head()
```

```python
m = titanic.loc[titanic['Sex'] == 'male']
fe = titanic.loc[titanic['Sex'] == 'female']
```

```python
print(m.count(), fe.count())
```

```python
plt.subplot(1,2,1)
plt.ylim([0, 400])
plt.hist(m['Pclass'])
plt.subplot(1,2,2)
plt.ylim([0, 400])
plt.hist(fe['Pclass'])
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.ylim([0, 120])
plt.title('남자 나이 분포')
plt.xlabel('나이')
plt.ylabel('인원수')
plt.hist(m['Age'])
plt.subplot(1, 2, 2)
plt.ylim([0, 120])
plt.hist(fe['Age'])
plt.title('여자 나이 분포')
plt.xlabel('나이')
plt.show()
```

```python
m_s = m.loc[m['Survived'] == 1]
fe_s = fe.loc[fe['Survived'] == 1]
```

```python
plt.subplot(1, 2, 1)
plt.ylim([0, 120])
plt.title('남자(생존자) 나이 분포')
plt.xlabel('나이')
plt.ylabel('인원수')
plt.hist(m_s['Age'])
plt.subplot(1, 2, 2)
plt.ylim([0, 120])
plt.title('여자(생존자) 나이 분포')
plt.xlabel('나이')
plt.hist(fe_s['Age'])
plt.show()
```

```python
m_d = m.loc[m['Survived'] == 0]
fe_d = fe.loc[fe['Survived'] == 0]
```

```python
plt.subplot(1, 2, 1)
plt.ylim([0, 120])
plt.title('남자(사망자) 나이 분포')
plt.xlabel('나이')
plt.ylabel('인원수')
plt.hist(m_d['Age'])
plt.subplot(1, 2, 2)
plt.ylim([0, 120])
plt.title('여자(사망자) 나이 분포')
plt.xlabel('나이')
plt.hist(fe_d['Age'])
plt.show()
```
